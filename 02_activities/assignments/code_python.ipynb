```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
df = pd.read_csv("/Users/jinrikim/Desktop/DSI/visualization/baby_names_male.csv")
```

```python
#filter rows with the name 'noah'
noah = df[df['Name'] == 'NOAH']
noah
```

```python
#filter rows from 1990
noah90 = noah[df['Year'] > 1989 ]
```

```python
noah90
```

```python
#Create a line chart to show the changing trend of the name Noah
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize = (7,5))
plt.plot(noah90['Year'],noah90['Frequency'], color = "darkcyan", linewidth = 2.5, linestyle= 'solid')

ax.set_xlabel('Year')
ax.set_ylabel('Frequency')
ax.set_title("The evolution of the popularity of the Name Noah(1990-2024)", fontsize = 14)
ax.text(1989.5,70,"38", transform =ax.transData, fontsize= 9)
ax.text(2020.5,905,"892", transform =ax.transData, fontsize= 9)

path='/Users/jinrikim/Desktop/DSI/visualization/02_activities/assignments'
filename = 'evolutino_noah.png'

plt.savefig(filename, dpi = 300)
```
